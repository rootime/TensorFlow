```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras

# Call back Class to stop when it reach to the target loss
loss_threshold = 0.000000001
class Model_CB(tf.keras.callbacks.Callback):  
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('loss') < loss_threshold):
      print(f"\n Loss is lower than ${loss_threshold}") 
      print("\n Stop Training")
      self.model.stop_training = True

callbacks = Model_CB()

# Data 
fmnist = tf.keras.datasets.fashion_mnist
(train_x, train_y), (test_x, test_y) = fmnist.load_data()

# Train Data Normalize
train_x, train_y = train_x/256, train_y/256

# Model
model = keras.models.Sequential(
    [ keras.layers.Flatten(), 
     keras.layers.Dense(128, activation = 'relu'),
     keras.layers.Dense(64, activation = 'relu'),
     keras.layers.Dense(32, activation = 'relu'),
     keras.layers.Dense(16, activation = 'relu'),
     keras.layers.Dense(10, activation = 'softmax')]
)
model.compile(optimizer = 'adam', 
              loss = 'sparse_categorical_crossentropy')


# Model Training
model.fit(train_x, train_y, epochs = 5, callbacks=[callbacks])
```

```python

# Evaluation
model.evaluate(test_x, test_y)

# Test Result
classification = model.predict(test_x)
print(classification[0])
```
